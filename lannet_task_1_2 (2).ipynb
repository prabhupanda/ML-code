```
#defining the function with input as a dataframe
def date_parsing(dataset):
  import numpy as np
  import pandas as pd
  # get the list of all date columns in the dataset & store it in a list
  # date columns have a datatypecof pandas object 
  cols=[]
  for i in range(len(dataset.columns)):
    cols.append(dataset.columns[i])     #stores all column headers in a list
  object_cols=[]
  for j in cols:                        #compares the dtypes of column in dataset to object type
    if dataset[j].dtypes== object:
      object_cols.append(j)
    else:
      pass
  #create a cache_dataframe for storing the object datatype columns
  cache_data=pd.DataFrame()
  #iterate through the object_cols list and populate the dataframe with all object comuns from dataset
  for y in range(len(object_cols)):
    cache_data[object_cols[y]]=dataset[object_cols[y]]
  
  #cache_data (dataframe) has all object type columns 
  #change all object_type entries to pandas_datetime format
  
  parsed_date=pd.DataFrame()
  for p in object_cols:
    parsed_date[p+str('_parsed')]=pd.to_datetime(cache_data[p],infer_datetime_format=True,errors='coerce',format='%d-%m-%Y')
    
  #parsed_data is populated with datetime columns only
  no_cols=[]
  diff_df=pd.DataFrame()
  for x in parsed_date.columns:
    no_cols.append(x)
  for a in range(len(no_cols)):
    for b in range(len(no_cols)):
      if a!=b and a<b: #computes the difference between two dates

        diff_df[str(no_cols[a])+'_&_'+str(no_cols[b])]= parsed_date[no_cols[b]]-parsed_date[no_cols[a]]
      else:
        pass
        
                                                                   
 
  return diff_df  
```

```
#dataset_link : https://drive.google.com/file/d/1Je7CtvN3maJ6EUmaljmWVbqlX0sXe4n0/view?usp=sharing
data_pearson=pd.read_csv('lannet_pearson.csv')
```

```
def pearson_coor(dataset):
  import numpy as np
  import pandas as pd
  from scipy.stats import pearsonr  #function to calculate pearson's correlation betn 2 columns
  X=dataset.iloc[:,:].values        #matrix of shape lxl ; where l = no. of attributes in dataset
  coor_matrix=np.zeros((X.shape[1],X.shape[1]))
  for i in range(X.shape[1]):       #to iterate through all columns & calculate pearsons coefficient
    for j in range(X.shape[1]):
      if i!=j and i<j:
        pear_col=pearsonr(X[:,i],X[:,j])
        coor_matrix[i,j]=pear_col[0]
      else:
        pass
  lists=[]
  #iterate through the whole matrix to search the index of coefficeint>0.85
  for p in range(coor_matrix.shape[0]):
    for s in range(coor_matrix.shape[1]):
      if coor_matrix[p,s]>0.85:
        lists.append([p,s])   #lists all the corelated columns indexes
  #Delete the correlated columns from the dataset by iterating through the lists (list)
  dataset_cols=dataset.columns
  dropping_cols=[]
  for a in range(len(lists)):
    dropping_cols.append(dataset_cols[lists[a][1]]) #names of columns to be dropped are stored in dropping_cols

  #Dropping the columns from Dataframe
  for b in dropping_cols:
    dataset=dataset.drop([b],axis=1)

  return dataset
```
