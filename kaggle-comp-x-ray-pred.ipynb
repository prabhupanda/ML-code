```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
import seaborn as sns
sns.set()
```

```python
train_gen=ImageDataGenerator(rescale=1./255,
                            horizontal_flip=True,
                            vertical_flip=False,
                            rotation_range=10,
                            
                            )
train_data=train_gen.flow_from_directory('/kaggle/input/dlai3-phase3/DLAI3_Phase3/DLAI3_Phase3',
                                        target_size=(512,512),
                                        color_mode='rgb',
                                        batch_size=32)
#training_images=tf.image.grayscale_to_rgb(images=train_data)
```

Importing the Validation Dataset (**sorting is required : shuffle=False )

```python
valid_gen=ImageDataGenerator(rescale=1./255,
                            horizontal_flip=False,
                            vertical_flip=False
                            )
valid_data=valid_gen.flow_from_directory('/kaggle/input/dlai3-phase3/VALIDATE',
                                        target_size=(512,512),
                                        color_mode='rgb',
                                        batch_size=32,
                                        shuffle=False)
```

```python
#creating a transfer learning model
#downloading a siutable model
xcp_framework=keras.applications.Xception(weights='imagenet',
                                      include_top=False,
                                      input_shape=(512,512,3))

for layer in xcp_framework.layers:
    layer.trainable=False
```

```python
#hyperparameters
# metrics=keras.metrics.CategoricalAccuracy
# optimizer=keras.optimizers.Adam
# loss=keras.losses.CategoricalCrossentropy


my_model=keras.models.Sequential()
my_model.add(xcp_framework)
my_model.add(keras.layers.Flatten())
my_model.add(keras.layers.BatchNormalization())
my_model.add(keras.layers.Dropout(0.3))
my_model.add(keras.layers.Dense(512,activation='relu'))
my_model.add(keras.layers.Dropout(0.3))
my_model.add(keras.layers.Dense(3,activation='softmax'))
my_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['categorical_accuracy']) 
```

```python
my_model.summary()
```

```python
tf.test.gpu_device_name()
```

```python
history=my_model.fit(train_data,batch_size=1024,epochs=5)
```

```python
n=len(history.history['loss'])
```

```python
plt.plot(np.arange(0,n),history.history['loss'])
plt.plot(np.arange(0,n),history.history['categorical_accuracy'])
```

```python
predict=my_model.predict(valid_data)
```
